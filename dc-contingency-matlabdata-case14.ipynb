```julia
# this function is to compute the scopf problem with one specific branch contingency
# the input is the dataset of the whole network and the contingency
# the output is solution of the problem if it has feasible, if it is not feasible the model will return "infeasible"
function run_dc_scopf_contingency(network,contingency::Int64)
    
    bus_number = length(network["bus"])
    gen_number = length(network["gen"])
    branch_number = length(network["branch"])
    # variable
    model_dc_scopf = Model(with_optimizer(Ipopt.Optimizer, tol=1e-9, print_level=0))
    @variable(model_dc_scopf, flow[1:branch_number])
    @variable(model_dc_scopf,voltage_phase[1:bus_number]) 
    @variable(model_dc_scopf, bus_p[1:bus_number],lower_bound=0)
    @variable(model_dc_scopf, gen_p[1:gen_number])
    @variable(model_dc_scopf, gen_cost[1:gen_number])



        
    beta_g = []
    for i in 1:gen_number
        gen_index = string(i)
        push!(beta_g,network["gen"][gen_index]["gen_bus"])
    end  

        bus_p0 = zeros(1,bus_number)
        for i in 1:gen_number
            gen_index = string(i)
            bus_name = network["gen"][gen_index]["gen_bus"]
            bus_produce = network["gen"][gen_index]["pg"]
            bus_p0[bus_name] = bus_produce
        end

    bus_d = zeros(1,bus_number)
    for i in 1:length(network["load"])
        load_index = string(i)
        bus_name = network["load"][load_index]["load_bus"]
        bus_demand = network["load"][load_index]["pd"]
        bus_d[bus_name] = bus_demand
    end

    ab_l = []
    for i in 1:branch_number
        abl = []
        branch_index = string(i)
        push!(abl,network["branch"][branch_index]["f_bus"])
        push!(abl,network["branch"][branch_index]["t_bus"])
        push!(ab_l,abl)
    end

    A_bl=zeros(bus_number,branch_number)
    from_bus = 0
    to_bus = 0
    for i in 1:branch_number
        from_bus = ab_l[i][1]
        to_bus = ab_l[i][2]
        A_bl[(i-1)*bus_number+from_bus] = -1
        A_bl[(i-1)*bus_number+to_bus] = 1
    end 

    flow_bus_matrix = A_bl*flow

    line_matrix = A_bl'*voltage_phase
    V0 = 1

    bus_p_max = zeros(1,bus_number)
    for i in 1:gen_number
        gen_index = string(i)
        bus_name = network["gen"][gen_index]["gen_bus"]
        bus_produce_max = network["gen"][gen_index]["pmax"]
        bus_p_max[bus_name] = bus_produce_max
    end
    
        bus_p_min = zeros(1,bus_number)
        for i in 1:gen_number
            gen_index = string(i)
            bus_name = network["gen"][gen_index]["gen_bus"]
            bus_produce_min = network["gen"][gen_index]["pmin"]
            bus_p_min[bus_name] = bus_produce_min
        end

    
   for i in 1:gen_number
       gen_index = string(i)
       bus_index = network["gen"][gen_index]["gen_bus"]
       @constraint(model_dc_scopf,gen_p[i] == bus_p[bus_index])
   end

    # value of objective function
    gen_cost_con=[]
    for i in 1:gen_number
        gen_index = string(i)
        push!(gen_cost_con,@constraint(model_dc_scopf,gen_cost[i] == network["gen"][gen_index]["cost"][1]*gen_p[i]^2+network["gen"][gen_index]["cost"][2]*gen_p[i]+network["gen"][gen_index]["cost"][3]))
    end    
    
    F_max = []
    for i in 1:branch_number
        branch_index = string(i)
        push!(F_max,network["branch"][branch_index]["rate_a"])
    end  

    # generation limit
    for i in 1:bus_number
        if i in beta_g
            @constraint(model_dc_scopf, bus_p[i] >= bus_p_min[i])
            @constraint(model_dc_scopf, bus_p[i] <= bus_p_max[i])
        else
            @constraint(model_dc_scopf, bus_p[i] == 0)
        end
    end

    # thermal limit
    for i in 1:branch_number
        @constraint(model_dc_scopf, flow[i] - F_max[i]<=0)
        @constraint(model_dc_scopf, -flow[i]- F_max[i]<=0)
    end 

    # Kirchhoff current law
    con1= []
    for i in 1:bus_number
        bus_index = string(i)
        push!(con1,@constraint(model_dc_scopf, flow_bus_matrix[i]- bus_d[i]== -bus_p[i] ))
    end

    #contingency : the power flow in contingency branch is zero
    @constraint(model_dc_scopf, flow[contingency] == 0)
    
    # Kirchhoff voltage law(without contingency branch)
    con2 = []
    for i in 1:branch_number
        if i==contingency
        else
            branch_index = string(i)
            push!(con2, @constraint(model_dc_scopf, (-V0^2/network["branch"][branch_index]["br_x"])*line_matrix[i] == flow[i]))
        end
    end
    
    # Setting the Reference Bus (Bus 1) angles to 0
    @constraint(model_dc_scopf, voltage_phase[1] == 0)


    # objective
    @objective(model_dc_scopf, Min,sum(gen_cost))
    
    #set_optimizer(model_dc_scopf, Ipopt.Optimizer,print_level=0)
    optimize!(model_dc_scopf) 
    
    status = termination_status(model_dc_scopf)
    if status == LOCALLY_SOLVED
        return(value.(sum(gen_cost)))
    else
        return("infeasible")
    end
    
    
end
```

```julia
# this function is to compute the scopf problem with all feasible branch contingency
# the input is the dataset of the whole network 
# the output is solution of the problem if it has feasible, if it is not feasible the model will return "infeasible"
# this function is very similar to "run_dc_scopf_contingency"
function run_dc_scopf(network)
    
    bus_number = length(network["bus"])
    gen_number = length(network["gen"])
    branch_number = length(network["branch"])
    contingency_number = length(network["branch"])

    model_dc_scopf = Model(with_optimizer(Ipopt.Optimizer))
    @variable(model_dc_scopf, flow[1:contingency_number,1:branch_number])
    @variable(model_dc_scopf,voltage_phase[1:contingency_number,1:bus_number]) 
    @variable(model_dc_scopf, bus_p[1:bus_number],lower_bound=0)
    @variable(model_dc_scopf, gen_p[1:gen_number])
    @variable(model_dc_scopf, gen_cost[1:gen_number])
    
    
    beta_g = []
    for i in 1:gen_number
        gen_index = string(i)
        push!(beta_g,network["gen"][gen_index]["gen_bus"])
    end  

    bus_p0 = zeros(1,bus_number)
    for i in 1:gen_number
        gen_index = string(i)
        bus_name = network["gen"][gen_index]["gen_bus"]
        bus_produce = network["gen"][gen_index]["pg"]
        bus_p0[bus_name] = bus_produce
    end

    bus_d = zeros(1,bus_number)
    for i in 1:length(network["load"])
        load_index = string(i)
        bus_name = network["load"][load_index]["load_bus"]
        bus_demand = network["load"][load_index]["pd"]
        bus_d[bus_name] = bus_demand
    end

    ab_l = []
    for i in 1:branch_number
        abl = []
        branch_index = string(i)
        push!(abl,network["branch"][branch_index]["f_bus"])
        push!(abl,network["branch"][branch_index]["t_bus"])
        push!(ab_l,abl)
    end

    A_bl=zeros(bus_number,branch_number)
    from_bus = 0
    to_bus = 0
    for i in 1:branch_number
        from_bus = ab_l[i][1]
        to_bus = ab_l[i][2]
        A_bl[(i-1)*bus_number+from_bus] = -1
        A_bl[(i-1)*bus_number+to_bus] = 1
    end 

    V0 = 1

    bus_p_max = zeros(1,bus_number)
    for i in 1:gen_number
        gen_index = string(i)
        bus_name = network["gen"][gen_index]["gen_bus"]
        bus_produce_max = network["gen"][gen_index]["pmax"]
        bus_p_max[bus_name] = bus_produce_max
    end
    
    bus_p_min = zeros(1,bus_number)
    for i in 1:gen_number
        gen_index = string(i)
        bus_name = network["gen"][gen_index]["gen_bus"]
        bus_produce_min = network["gen"][gen_index]["pmin"]
        bus_p_min[bus_name] = bus_produce_min
    end
    
        
    F_max = []
    for i in 1:branch_number
        branch_index = string(i)
        push!(F_max,network["branch"][branch_index]["rate_a"])
    end  
    
    for i in 1:gen_number
        gen_index = string(i)
        bus_index = network["gen"][gen_index]["gen_bus"]
        @constraint(model_dc_scopf,gen_p[i] == bus_p[bus_index])
    end

    gen_cost_con=[]
    for i in 1:gen_number
        gen_index = string(i)
        push!(gen_cost_con,@constraint(model_dc_scopf,gen_cost[i] == network["gen"][gen_index]["cost"][1]*gen_p[i]^2+network["gen"][gen_index]["cost"][2]*gen_p[i]+network["gen"][gen_index]["cost"][3]))
    end    


    for i in 1:bus_number
        if i in beta_g
            @constraint(model_dc_scopf, bus_p[i] >= bus_p_min[i])
            @constraint(model_dc_scopf, bus_p[i] <= bus_p_max[i])
        else
            @constraint(model_dc_scopf, bus_p[i] == 0)
        end
    end
    
    # consraints under all contingency 
    for i in 1:contingency_number
        
        for j in 1:branch_number
            @constraint(model_dc_scopf, flow[i,j] - F_max[j]<=0)
            @constraint(model_dc_scopf, -flow[i,j]- F_max[j]<=0)
        end

        for j in 1:bus_number
            f_m = A_bl*flow[i,:]
            @constraint(model_dc_scopf, f_m[j] - bus_d[j]== -bus_p[j] )
        end
        
        l_m = A_bl'*voltage_phase[i,:]
        
        if i in infeasible_contingency
            for j in 1:branch_number
                branch_index = string(i)
                @constraint(model_dc_scopf, (-V0^2/network["branch"][branch_index]["br_x"])*l_m[j] == flow[i,j])
            end           
        else   
            for j in 1:branch_number
                if j==i
                # the power flow in contingency branch is zero
                    @constraint(model_dc_scopf, flow[i,j] == 0)
                else
                    branch_index = string(i)
                    @constraint(model_dc_scopf, (-V0^2/network["branch"][branch_index]["br_x"])*l_m[j] == flow[i,j])
                end
            end
        end
    
    end    

    # objective
    @objective(model_dc_scopf, Min,sum(gen_cost))
    
    optimize!(model_dc_scopf) 
    
    status = termination_status(model_dc_scopf)
    if status == LOCALLY_SOLVED
        return(value.(gen_p))
    else
        return("infeasible")
    end
    
end
```

```julia
# this function is to find all the violation branch under all contingency
# the input is the dataset of the whole network
# the output is the number of violation contingencies, the index of these branches, their violation value and the corresopnding lagrange multiplier
# this can be used to compute all the sub problems for a grid
function benders_subproblem_test(network)
    contingency_violation = []
    lagrangian_multiplier_set = []
    violate_branch = []
    
    # scan all contingency 
    for contingency in 1:length(network["branch"])
    
        bus_number = length(network["bus"])
        gen_number = length(network["gen"])
        branch_number = length(network["branch"])
        # model name
        model_dc_slack = Model()
        # model variables
        # slack variables
        @variable(model_dc_slack, slack1[1:branch_number],lower_bound=0)
        @variable(model_dc_slack, slack2[1:branch_number],lower_bound=0)
        @variable(model_dc_slack, slack3[1:bus_number],lower_bound=0)
        @variable(model_dc_slack, slack4[1:bus_number],lower_bound=0)
        @variable(model_dc_slack, flow[1:branch_number])
        @variable(model_dc_slack,voltage_phase[1:bus_number]) 

    
        # the parameters that needed in the model
        # max thermal limits for each line
        F_max = []
        for i in 1:branch_number
            branch_index = string(i)
            push!(F_max,network_con["branch"][branch_index]["rate_a"])
        end  

        beta_g = []
        for i in 1:gen_number
            gen_index = string(i)
           push!(beta_g,network["gen"][gen_index]["gen_bus"])
        end  
        # given value of generation
        bus_p0 = zeros(1,bus_number)
        for i in 1:gen_number
            gen_index = string(i)
            bus_name = network["gen"][gen_index]["gen_bus"]
            bus_produce = network["gen"][gen_index]["pg"]
            bus_p0[bus_name] = bus_produce
        end
        # power demand for each bus
        bus_d = zeros(1,bus_number)
        for i in 1:length(network["load"])
            load_index = string(i)
            bus_name = network["load"][load_index]["load_bus"]
            bus_demand = network["load"][load_index]["pd"]
            bus_d[bus_name] = bus_demand
        end

        ab_l = []
        for i in 1:branch_number
            abl = []
            branch_index = string(i)
            push!(abl,network["branch"][branch_index]["f_bus"])
            push!(abl,network["branch"][branch_index]["t_bus"])
            push!(ab_l,abl)
        end

        A_bl=zeros(bus_number,branch_number)
        from_bus = 0
        to_bus = 0
        for i in 1:branch_number
            from_bus = ab_l[i][1]
            to_bus = ab_l[i][2]
            A_bl[(i-1)*bus_number+from_bus] = -1
            A_bl[(i-1)*bus_number+to_bus] = 1
        end 
        

        flow_bus_matrix = A_bl*flow

        V0 = 1
        line_matrix = A_bl'*voltage_phase
        
        # flow constrain
        flow_max_con = []
        flow_min_con = []
        for i in 1:branch_number
            push!(flow_max_con,@constraint(model_dc_slack, flow[i]-slack1[i] <= F_max[i]))
            push!(flow_min_con,@constraint(model_dc_slack, - F_max[i] <= flow[i] + slack2[i] ))
        end 
    
        #Kirchhoff current law
        power_balance = []
        for i in 1:bus_number
            bus_index = string(i)
            push!(power_balance,@constraint(model_dc_slack, flow_bus_matrix[i]- bus_d[i]+ slack3[i] - slack4[i]== -bus_p0[i] ))
        end

        # Setting the Reference Bus (Bus 1) angles to 0
        @constraint(model_dc_slack, voltage_phase[1] == 0)

        #contingency : the power flow in contingency branch is zero
        @constraint(model_dc_slack, flow[contingency] == 0)
        
        # Kirchhoff voltage law(without contingency branch)
        voltage_law = []
        for i in 1:(contingency-1)
            branch_index = string(i)
            push!(voltage_law, @constraint(model_dc_slack, (-V0^2/network_con["branch"][branch_index]["br_x"])*line_matrix[i] == flow[i]))
        end 
        for i in (contingency+1):branch_number
            branch_index = string(i)
            push!(voltage_law, @constraint(model_dc_slack, (-V0^2/network_con["branch"][branch_index]["br_x"])*line_matrix[i] == flow[i]))
        end 



        # objective function: minimize the all the violation value
        violation = @objective(model_dc_slack, Min, sum(sum(sum(slack1[i] for i in [1:branch_number])+ sum(slack2[j] for j in [1:branch_number]))+sum(sum(slack3[m] for m in [1:bus_number])+sum(slack4[n] for n in [1:bus_number]))))

        
        # set "Ipopt" as solver
        set_optimizer(model_dc_slack, Xpress.Optimizer)
        optimize!(model_dc_slack)
    

    #lagrangian_multiplier 
    lagrangian_multiplier =  -dual.(power_balance)

        # check violation value is larger than 0 or not
        if value.(violation)>10e-6
             push!(lagrangian_multiplier_set,lagrangian_multiplier)
             push!(violate_branch,contingency)
             push!(contingency_violation,value.(violation))
        end         
    end
    num = length(violate_branch)
    # final output
    return (num,violate_branch,lagrangian_multiplier_set,contingency_violation)

end
    
```

```julia
# this function is to find all the power flow violation branch for a specific contingency
# the input is the dataset of the whole network and a contingency
# the output is to tell whether the critical contingency is power flow violation contingency or not
# this can only be used to check one contingency for each time
# this function is almost the same with the above one
function benders_subproblem(network, contingency)
    contingency_violation = []
    lagrangian_multiplier_set = []
    violate_branch = []
    
    
    bus_number = length(network["bus"])
    gen_number = length(network["gen"])
    branch_number = length(network["branch"])
    
    model_dc_fpv = Model()
    @variable(model_dc_fpv, slack1[1:branch_number],lower_bound=0)
    @variable(model_dc_fpv, slack2[1:branch_number],lower_bound=0)
    @variable(model_dc_fpv, slack3[1:bus_number],lower_bound=0)
    @variable(model_dc_fpv, slack4[1:bus_number],lower_bound=0)
    @variable(model_dc_fpv, flow[1:branch_number])
    @variable(model_dc_fpv,voltage_phase[1:bus_number]) 

    
    #F_max = 2
    F_max = []
    for i in 1:branch_number
        branch_index = string(i)
        push!(F_max,network_con["branch"][branch_index]["rate_a"])
    end  

    beta_g = []
    for i in 1:gen_number
        gen_index = string(i)
        push!(beta_g,network["gen"][gen_index]["gen_bus"])
    end  

    bus_p0 = zeros(1,bus_number)
    for i in 1:gen_number
        gen_index = string(i)
        bus_name = network["gen"][gen_index]["gen_bus"]
        bus_produce = network["gen"][gen_index]["pg"]
        bus_p0[bus_name] = bus_produce
    end

    bus_d = zeros(1,bus_number)
    for i in 1:length(network["load"])
        load_index = string(i)
        bus_name = network["load"][load_index]["load_bus"]
        bus_demand = network["load"][load_index]["pd"]
        bus_d[bus_name] = bus_demand
    end

    ab_l = []
    for i in 1:branch_number
        abl = []
        branch_index = string(i)
        push!(abl,network["branch"][branch_index]["f_bus"])
        push!(abl,network["branch"][branch_index]["t_bus"])
        push!(ab_l,abl)
    end

    A_bl=zeros(bus_number,branch_number)
    from_bus = 0
    to_bus = 0
    for i in 1:branch_number
        from_bus = ab_l[i][1]
        to_bus = ab_l[i][2]
        A_bl[(i-1)*bus_number+from_bus] = -1
        A_bl[(i-1)*bus_number+to_bus] = 1
    end 
        

    flow_bus_matrix = A_bl*flow



    V0 = 1
    line_matrix = A_bl'*voltage_phase

    flow_max_con = []
    flow_min_con = []
    for i in 1:branch_number
        push!(flow_max_con,@constraint(model_dc_fpv, flow[i]-slack1[i] <= F_max[i]))
        push!(flow_min_con,@constraint(model_dc_fpv, - F_max[i] <= flow[i] + slack2[i] ))
    end 
    
    power_balance = []
    for i in 1:bus_number
        bus_index = string(i)
        push!(power_balance,@constraint(model_dc_fpv, flow_bus_matrix[i]- bus_d[i]+ slack3[i] - slack4[i]== -bus_p0[i] ))
    end

    # Setting the Reference Bus (Bus 1) angles to 0
    @constraint(model_dc_fpv, voltage_phase[1] == 0)

    # set contingency 
    @constraint(model_dc_fpv, flow[contingency] == 0)
    # Kirchhoff voltage law
    voltage_law = []
    for i in 1:(contingency-1)
        branch_index = string(i)
        push!(voltage_law, @constraint(model_dc_fpv, (-V0^2/network_con["branch"][branch_index]["br_x"])*line_matrix[i] == flow[i]))
    end 

    for i in (contingency+1):branch_number
        branch_index = string(i)
        push!(voltage_law, @constraint(model_dc_fpv, (-V0^2/network_con["branch"][branch_index]["br_x"])*line_matrix[i] == flow[i]))
    end 



    # objective
    violation = @objective(model_dc_fpv, Min, sum(sum(sum(slack1[i] for i in [1:branch_number])+ sum(slack2[j] for j in [1:branch_number]))+sum(sum(slack3[m] for m in [1:bus_number])+sum(slack4[n] for n in [1:bus_number]))))

    set_optimizer(model_dc_fpv, Xpress.Optimizer)
    optimize!(model_dc_fpv)
    

    #lagrangian_multiplier 
    lagrangian_multiplier =  -dual.(power_balance)

    # check violation value is larger than 0 or not
    if value.(violation)>10e-8
        return(contingency,lagrangian_multiplier,value.(violation))
    else
        return("not violation")
    end
end
    
```

```julia
# this function is add benders cut to the master problem for a specific contingency violation
# the input is the dataset of the whole network , a contingency, violation value ans lagrangian multiplier
# the output is solution of basic dcopf model with a benders cut
# this can only be used to one cut for each time
function run_dc_opf_with_benderscut(network,lagrangian_multiplier_set_list::Array,contingency_violation_list::Array,violate_branch_number::Int64)
    
    bus_number = length(network["bus"])
    gen_number = length(network["gen"])
    branch_number = length(network["branch"])

    model_dc_benders_cut = Model(with_optimizer(Ipopt.Optimizer))
    # model variable
    @variable(model_dc_benders_cut, flow[1:branch_number])
    @variable(model_dc_benders_cut,voltage_phase[1:bus_number]) 
    @variable(model_dc_benders_cut, bus_p[1:bus_number],lower_bound=0)
    @variable(model_dc_benders_cut, gen_p[1:gen_number])
    @variable(model_dc_benders_cut, gen_cost[1:gen_number])
    # this variable is used to generated benders cut 
    @variable(model_dc_benders_cut, bus_add_cut[1:bus_number])

    beta_g = []
    for i in 1:gen_number
        gen_index = string(i)
        push!(beta_g,network["gen"][gen_index]["gen_bus"])
    end  

    bus_p0 = zeros(1,bus_number)
    for i in 1:gen_number
        gen_index = string(i)
        bus_name = network["gen"][gen_index]["gen_bus"]
        bus_produce = network["gen"][gen_index]["pg"]
        bus_p0[bus_name] = bus_produce
    end

    bus_d = zeros(1,bus_number)
    for i in 1:length(network["load"])
        load_index = string(i)
        bus_name = network["load"][load_index]["load_bus"]
        bus_demand = network["load"][load_index]["pd"]
        bus_d[bus_name] = bus_demand
    end

    ab_l = []
    for i in 1:branch_number
        abl = []
        branch_index = string(i)
        push!(abl,network["branch"][branch_index]["f_bus"])
        push!(abl,network["branch"][branch_index]["t_bus"])
        push!(ab_l,abl)
    end

    A_bl=zeros(bus_number,branch_number)
    from_bus = 0
    to_bus = 0
    for i in 1:branch_number
        from_bus = ab_l[i][1]
        to_bus = ab_l[i][2]
        A_bl[(i-1)*bus_number+from_bus] = -1
        A_bl[(i-1)*bus_number+to_bus] = 1
    end 

    flow_bus_matrix = A_bl*flow

    line_matrix = A_bl'*voltage_phase
    V0 = 1

    bus_p_max = zeros(1,bus_number)
    for i in 1:gen_number
        gen_index = string(i)
        bus_name = network["gen"][gen_index]["gen_bus"]
        bus_produce_max = network["gen"][gen_index]["pmax"]
        bus_p_max[bus_name] = bus_produce_max
    end
    
        bus_p_min = zeros(1,bus_number)
        for i in 1:gen_number
            gen_index = string(i)
            bus_name = network["gen"][gen_index]["gen_bus"]
            bus_produce_min = network["gen"][gen_index]["pmin"]
            bus_p_min[bus_name] = bus_produce_min
        end

    #bender cut
    benders1 = []
    for i in 1:bus_number
        push!(benders1,@constraint(model_dc_benders_cut,(bus_p[i]-bus_p0[i])*lagrangian_multiplier_set_list[violate_branch_number][i] ==bus_add_cut[i]))
    end
    
     benders_cut_add = @constraint(model_dc_benders_cut, sum(bus_add_cut)+contingency_violation_list[violate_branch_number]<=0)

    # total generation cost
    for i in 1:gen_number
        gen_index = string(i)
        bus_index = network["gen"][gen_index]["gen_bus"]
        @constraint(model_dc_benders_cut,gen_p[i] == bus_p[bus_index])
    end

    gen_cost_con=[]
    for i in 1:gen_number
        gen_index = string(i)
        push!(gen_cost_con,@constraint(model_dc_benders_cut,gen_cost[i] == network["gen"][gen_index]["cost"][1]*gen_p[i]^2+network["gen"][gen_index]["cost"][2]*gen_p[i]+network["gen"][gen_index]["cost"][3]))
    end    
    

     F_max = []
     for i in 1:branch_number
         branch_index = string(i)
         push!(F_max,network_con["branch"][branch_index]["rate_a"])
     end  

    # the following constraints are the constraints in DCOPF model
    
     # generation level limit
     for i in 1:bus_number
         if i in beta_g
             @constraint(model_dc_benders_cut, bus_p[i] >= bus_p_min[i])
             @constraint(model_dc_benders_cut, bus_p[i] <= bus_p_max[i])
         else
             @constraint(model_dc_benders_cut, bus_p[i] == 0)
         end
     end

    # thermal limits
     for i in 1:branch_number
         @constraint(model_dc_benders_cut, flow[i] - F_max[i]<=0)
         @constraint(model_dc_benders_cut, -flow[i]- F_max[i]<=0)
     end 

    # KCL
     con1= []
     for i in 1:bus_number
         bus_index = string(i)
         push!(con1,@constraint(model_dc_benders_cut, flow_bus_matrix[i]- bus_d[i]== -bus_p[i] ))
     end

    # KVL
    con2 = []
    for i in 1:branch_number
        branch_index = string(i)    
        push!(con2,@constraint(model_dc_benders_cut, (-V0^2/network["branch"][branch_index]["br_x"])*line_matrix[i] == flow[i]))
    end 
 
    # Setting the Reference Bus (Bus 1) angles to 0
    @constraint(model_dc_benders_cut, voltage_phase[1] == 0)


    # objective function
    @objective(model_dc_benders_cut, Min,sum(gen_cost))
    optimize!(model_dc_benders_cut)
    
    status = termination_status(model_dc_benders_cut)
    solved_branch = 0
    if status == LOCALLY_SOLVED
        solved_branch = violate_branch_number
    else
        solved_branch = 0
    end
    
    return (value.(gen_p),value.(sum(gen_cost)))
end
```

```julia
# this function is add all benders cuts to the master problem  
# the input is the dataset of the whole network , the critical contingency list, violation value list , lagrangian multiplier list, value of given bus
# the output is solution of basic dcopf model with a benders cut
# this function is almost the same with "run_dc_opf_with_benderscut",but it can add all cuts 
function run_dc_opf_all_benderscut(network,lagrangian_multiplier_set_list::Array,contingency_violation_list::Array,bus_p0_list::Array)
    
    bus_number = length(network["bus"])
    gen_number = length(network["gen"])
    branch_number = length(network["branch"])

    model_dc_benders_cut = Model(with_optimizer(Ipopt.Optimizer))
    @variable(model_dc_benders_cut, flow[1:branch_number])
    @variable(model_dc_benders_cut,voltage_phase[1:bus_number]) 
    @variable(model_dc_benders_cut, bus_p[1:bus_number],lower_bound=0)
    @variable(model_dc_benders_cut, gen_p[1:gen_number])
    @variable(model_dc_benders_cut, gen_cost[1:gen_number])
    @variable(model_dc_benders_cut, bus_add_cut[1:bus_number])



    beta_g = []
    for i in 1:gen_number
        gen_index = string(i)
        push!(beta_g,network["gen"][gen_index]["gen_bus"])
    end  

    bus_d = zeros(1,bus_number)
    for i in 1:length(network["load"])
        load_index = string(i)
        bus_name = network["load"][load_index]["load_bus"]
        bus_demand = network["load"][load_index]["pd"]
        bus_d[bus_name] = bus_demand
    end

    ab_l = []
    for i in 1:branch_number
        abl = []
        branch_index = string(i)
        push!(abl,network["branch"][branch_index]["f_bus"])
        push!(abl,network["branch"][branch_index]["t_bus"])
        push!(ab_l,abl)
    end

    A_bl=zeros(bus_number,branch_number)
    from_bus = 0
    to_bus = 0
    for i in 1:branch_number
        from_bus = ab_l[i][1]
        to_bus = ab_l[i][2]
        A_bl[(i-1)*bus_number+from_bus] = -1
        A_bl[(i-1)*bus_number+to_bus] = 1
    end 

    flow_bus_matrix = A_bl*flow

    line_matrix = A_bl'*voltage_phase
    V0 = 1

    bus_p_max = zeros(1,bus_number)
    for i in 1:gen_number
        gen_index = string(i)
        bus_name = network["gen"][gen_index]["gen_bus"]
        bus_produce_max = network["gen"][gen_index]["pmax"]
        bus_p_max[bus_name] = bus_produce_max
    end
    
    bus_p_min = zeros(1,bus_number)
    for i in 1:gen_number
        gen_index = string(i)
        bus_name = network["gen"][gen_index]["gen_bus"]
        bus_produce_min = network["gen"][gen_index]["pmin"]
        bus_p_min[bus_name] = bus_produce_min
    end

    #adding all bender cuts to the basic DCOPF model
    #iteration = length(effective_violate_branch_list)
    for i in 1:length(contingency_violation_list)
        vio_num = length(contingency_violation_list[i])
        for j in 1:vio_num
            #index= effective_violate_branch_list[i][j]
            @constraint(model_dc_benders_cut, sum((bus_p[b]-bus_p0_list[i][b])*lagrangian_multiplier_set_list[i][j][b] for b in 1:bus_number)+contingency_violation_list[i][j]<=0)
        end
    end 
    
   for i in 1:gen_number
       gen_index = string(i)
       bus_index = network["gen"][gen_index]["gen_bus"]
       @constraint(model_dc_benders_cut,gen_p[i] == bus_p[bus_index])
   end

   gen_cost_con=[]
   for i in 1:gen_number
       gen_index = string(i)
       push!(gen_cost_con,@constraint(model_dc_benders_cut,gen_cost[i] == network["gen"][gen_index]["cost"][1]*gen_p[i]^2+network["gen"][gen_index]["cost"][2]*gen_p[i]+network["gen"][gen_index]["cost"][3]))
   end    
    
    F_max = []
    for i in 1:branch_number
        branch_index = string(i)
        push!(F_max,network_con["branch"][branch_index]["rate_a"])
    end  

    for i in 1:bus_number
        if i in beta_g
            @constraint(model_dc_benders_cut, bus_p[i] >= bus_p_min[i])
            @constraint(model_dc_benders_cut, bus_p[i] <= bus_p_max[i])
        else
            @constraint(model_dc_benders_cut, bus_p[i] == 0)
        end
    end

    for i in 1:branch_number
        @constraint(model_dc_benders_cut, flow[i] - F_max[i]<=0)
        @constraint(model_dc_benders_cut, -flow[i]- F_max[i]<=0)
    end 

    con1= []
    for i in 1:bus_number
        bus_index = string(i)
        push!(con1,@constraint(model_dc_benders_cut, flow_bus_matrix[i]- bus_d[i]== -bus_p[i] ))
    end

    con2 = []
    for i in 1:branch_number
        branch_index = string(i)    
        push!(con2,@constraint(model_dc_benders_cut, (-V0^2/network["branch"][branch_index]["br_x"])*line_matrix[i] == flow[i]))
    end 

    # Setting the Reference Bus (Bus 1) angles to 0
    @constraint(model_dc_benders_cut, voltage_phase[1] == 0)


    # objective
    @objective(model_dc_benders_cut, Min,sum(gen_cost))

    optimize!(model_dc_benders_cut) 
    #return value.(gen_p)
    return (value.(gen_p),value.(voltage_phase))
end
```

```julia
# There are threes step in this function
# the first step is to filter the contingency ,since some contingency cannot feasible for the basic SCOPF model
# the second step is to generated the benders cut from the filtered contingency and sovle the dcopf model with benders cut
# the third step is to check all the violation contingency and  generate the related vioaltion list 
# the output is to new dataset for the violation contingency if there still exists violation contingency
# the output is to tell that the optimal solution satisfied under all contingency if there is not violation contingency
# it include all the data that can be used for the next iteratio 
function check_recompute(lagrangian_multiplier_set_list0,violate_branch_list0,contingency_violation_list0,bus_p0)
    violate_list = []
    lagrangian_multiplier_set_list2 = []
    contingency_violation_list2 = []
    
    for i in 1:length(violate_branch_list0)
        if run_dc_scopf_contingency(case_data,violate_branch_list0[i]) == "infeasible"
        else
            push!(violate_list,i)
        end
    end
    println(violate_list)


    for i in violate_list
        push!(lagrangian_multiplier_set_list2,lagrangian_multiplier_set_list0[i])
        push!(contingency_violation_list2,contingency_violation_list0[i])
    end

    # compute the master problem with adding all benders cut
    network_basesolution = deepcopy(case_data)
    push!(bus_p0_list,bus_p0)
    push!(lm_list,lagrangian_multiplier_set_list2)
    push!(cv_list,contingency_violation_list2)
    push!(v_list,violate_list)

    # get the new value of power generation
    new = run_dc_opf_all_benderscut(case_data,lm_list,cv_list,bus_p0_list)
    new_result = collect(new)
    new_gen_p = new_result[1]
    new_voltage = new_result[2]
    println(new_gen_p)
    println(new_voltage)
    
    # use the new value to compute the sub problem
    network_con_revised = deepcopy(case_data)
    for i in 1:length(network_con_revised["gen"])
        gen_index = string(i)
        network_con_revised["gen"][gen_index]["pg"] = new_gen_p[i]
    end
    
    bus_p00 = zeros(1,bus_number)
    for i in 1:length(network_con_revised["gen"])
        gen_index = string(i)
        bus_name = network_con_revised["gen"][gen_index]["gen_bus"]
        bus_produce = new_gen_p[i]
        bus_p00[bus_name] = bus_produce
    end

    # find the all violation contingency with new generation value
    violation_details = benders_subproblem_test(network_con_revised)
    vd = collect(violation_details)
    violation_number = vd[1]
    violate_branch_list = vd[2]
    lagrangian_multiplier_set_list = vd[3]
    contingency_violation_list = vd[4]

    # find the effective violation contingency 
    violate_list = []
    for i in 1:length(violate_branch_list)
        if run_dc_scopf_contingency(case_data,violate_branch_list[i]) == "infeasible"
        else
            push!(violate_list,i)
        end
    end

    # print the result
    if length(violate_list) == 0
        return("we find the optimal solution to satisfy all branch contingencies for power flow violation")
    else
        info(LOGGER, "we DO NOT the optimal solution to satisfy all branch contingencies")
        println("viloation number:$(violation_number)")
        println("violate branch list:$(violate_branch_list)")
        println("contingency violation value list:$(contingency_violation_list)")
        return(lagrangian_multiplier_set_list,violate_branch_list,contingency_violation_list,bus_p00)        
    end
  
end
```
